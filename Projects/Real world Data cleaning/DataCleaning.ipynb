```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dirty_car_price_data.csv")
df.dropna(inplace = True)
df.head(10)
```

```python
def brandname(x):
    a = x.split()
    return  a[0].title()
    
def Carname(x):
    ind = x.index(" ")
    return  x[ind+1:].title()

def Conv(x):
    x = str(x)
    p = x.split();
    if(len(p)<=1):
        return x;
    else:
            if(p[1] == "Lakh"):
                return str(round(float(p[0])*100000,1))
            elif(p[1] == "Crore"):
                return str(round(float(p[0])*10000000,1))
       
```

```python
df["Brand"] = df['car_name'].apply(brandname)
df['car_name'] = df['car_name'].apply(Carname)
```

```python
df["prices_in_rup"] = df["prices_in_rup"].str.replace(",","")
df['prices_in_rup'] = df['prices_in_rup'].apply(Conv)
```

```python
df['prices_in_rup'] = df['prices_in_rup'].astype("float")
df
```

```python
def rem_kms(x):
    slt = x.split()
    return slt[0]
df['kms_driven'] = df['kms_driven'].apply(rem_kms)
df['kms_driven'] = df['kms_driven'].astype("int")
df
```

```python
def rem_cc(x):
    slt = x.split()
    return slt[0]
    
df['engine'] = df['engine'].apply(rem_cc)
df['engine'] = df['engine'].astype("int")
df
```

```python
def rem_seat(x):
    slt = x.split()
    return slt[0]
    
df['Seats'] = df['Seats'].apply(rem_seat)
df['Seats'] = df['Seats'].astype("int")
df
```

```python
df["manufacture"] = pd.to_datetime(df["manufacture"])
df
```
