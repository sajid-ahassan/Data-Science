```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# lineplot

```python
y1 = sns.load_dataset("penguins")
sns.lineplot(x = "bill_depth_mm",y ="bill_length_mm",data = y1,hue = "species",size = 10,
             palette = "Accent",markers = ["o",">"],legend = "auto")
plt.grid()
# hue = respect to(ex: male,female)
# palette = design
# markers = pevot marker
```

# Bar plot

```python
y1 = sns.load_dataset("penguins")
y2 = y1.head(15)
sns.set(style = "darkgrid")
order = ["Dream","Biscoe","Torgersen"]
sns.barplot(x = "island",y = "flipper_length_mm",data = y1,hue = "sex",order = order
           ,hue_order=["Female","Male"],palette = "icefire",saturation = 0.5
           ,alpha = 0.9,ci = 100,errcolor = "r",capsize = 0.2)


# For horigontal bar
# sns.barplot(x = "bill_depth_mm",y = "flipper_length_mm",data = y2,hue = "sex"
#            ,hue_order=["Female","Male"],orient = "h") # Both have to numarical values. String will not work

plt.tight_layout()
plt.show()

# hue = respect to(ex: male,female)
# palette = design
# order = custom order of bars
# orient =  vertical or horigontal
# hue_order = custom order of hue order
# saturation = control saturation
# ci = error bar length(0-100)
# errcolor = color of error bar
# capsize = size of horizontal error bar
# alpha = opasity
```

# Scatter plot

```python
y1 = sns.load_dataset("penguins")
y2 = y1.head(15)
sns.set(style = "darkgrid")
sns.scatterplot(x = "bill_depth_mm",y = "flipper_length_mm",data = y2,
                hue = "sex",style = "sex"
           ,hue_order=["Female","Male"],palette = "icefire"
           ,alpha = 0.9)
# hue = respect to(ex: male,female)
# palette = design
# order = custom order of bars
# orient =  vertical or horigontal
# hue_order = custom order of hue order
# alpha = opasity
```

# Box plot

```python
y1 = sns.load_dataset("penguins")
y2 = y1.head(15)
sns.set(style = "whitegrid")
sns.boxplot(x = "island",y = "flipper_length_mm",data = y1,hue = "sex",
            hue_order=["Female","Male"],palette = "icefire",showmeans = True,
           meanprops = {"marker":">","markeredgecolor":"b","markersize":7})
# hue = respect to(ex: male,female)
# palette = design
# order = custom order of bars
# orient =  vertical or horigontal
# hue_order = custom order of hue order
```

# Histogram

```python
y1 = sns.load_dataset("penguins")
sns.displot(y1["bill_length_mm"],bins = [30,35,40,45,50,55,60,65],kde = True,rug = True,color = "b")
plt.grid()
plt.title("Demo Histogram")
plt.show()
# bins = custom range
# kde = to add lineplot
# rug = dansity
```

# Heatmap

```python
arr = np.linspace(1,50,20).astype(np.int32)
edt = {"fontsize":10,"color":"b"}
arr = arr.reshape((4,5))
sn = sns.heatmap(arr,vmin = 0,vmax = 60,cmap ="PuOr",annot = True,
                 annot_kws = edt,linewidth = 3,linecolor ='r',
                 cbar = False)
sn.set(xlabel = "Python",ylabel = "java",title = "Nothing")
plt.show()

#vmin = color bar min value
#vmax = color bar max value
#cmap = theam
# annot = text in the box
# annot_kws =  text editing
# linewidth = adding space between boxs
#linecolor = space color
# cbar = show bar or not show
# sn.set() = to set labels
```
