```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df.isnull().sum()
df.dropna(axis = 0)
```

```python
df['age'] = df['age'].fillna(df['age'].mean()).astype(int)
df
```

```python
df.duplicated(subset = ['name','age','email'])
```

```python
df =df.ffill()
def func(x):
    if(x<=26): return 'adult'
    elif(x<30): return 'more adult'
    else: return 'old'
df['Experience'] = df['age'].apply(func)
df
```

```python
dct = {'UK':'United kingdom'}
df['country'] =df['country'].replace(dct)
df
```

```python
df.ffill()
df2 = df.sort_values(['age','id']).copy()
df2.reset_index(drop = True,inplace = True)
# df2.sort_index()
df2['Priopity'] = df2['age'].rank(ascending = True,method = 'dense')
df2
```

```python
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
df
```

```python

df.groupby("Department")["Salary"].agg(avg = "mean",mx = "max",mn = "min")
df.sort_values("TeamAvg",inplace = True,ascending = False)
df["TeamAvg"] = df.groupby("Department")["Salary"].transform("mean")
df["Salary"]
df.groupby("Team").filter(lambda x: x["Salary"].mean() > 80)
df.groupby("Department").min()
```
